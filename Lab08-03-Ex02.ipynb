## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

Field คือ ตัวแปรที่ถูกประกาศภายในคลาสหรือ struct และเป็นส่วนหนึ่งของข้อมูลของอ็อบเจ็กต์

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

Field ใช้สำหรับเก็บค่าหรือสถานะของอ็อบเจ็กต์ที่สร้างจากคลาส

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

class Example
{
    public int number; // ประกาศ Field ชนิด int
    private string name; // ประกาศ Field ชนิด string ที่เป็น private
}


4. Field และ Variable แตกต่างกันอย่างไร


Field เป็นตัวแปรที่อยู่ภายในคลาสหรือ struct และใช้เพื่อเก็บสถานะของอ็อบเจ็กต์

Variable เป็นตัวแปรทั่วไปที่สามารถอยู่ภายในเมธอด หรือบล็อกของโค้ดได้

5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

ควรใช้ Field เมื่อค่าต้องการถูกเก็บเป็นสถานะของอ็อบเจ็กต์

ควรใช้ Variable ภายในเมธอดเมื่อต้องการเก็บค่าชั่วคราวที่ใช้ภายในเมธอดนั้น
